```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/StockDataHDFCBANK.csv")
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
display(df.head())
```

Converting the 'Date' column to datetime objects to enable time-based plotting.


```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df
```

Checking for missing values in the DataFrame.



```python
print("Missing values before handling:")
print(df.isnull().sum())
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df['Date'], df['Close'], label='Close Price', color='blue')

ax.set_title('HDFC BANK Closing Price Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Close Price')

plt.xticks(rotation=45)
ax.grid(True)

x = np.arange(len(df))
y = df['Close'].values

m, b = np.polyfit(x, y, 1)
ax.plot(df['Date'], m*x + b, color='red', linestyle='--', label=f'Linear Trend (slope={m:.4f})')

rolling_mean = df['Close'].rolling(window=50, min_periods=1).mean()
ax.plot(df['Date'], rolling_mean, color='green', linestyle='-', label='50-Period Rolling Average')

ax.legend()
plt.tight_layout()
plt.show()
```
